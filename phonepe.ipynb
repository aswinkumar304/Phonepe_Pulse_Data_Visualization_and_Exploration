```python
# aggregated_transaction

path1 = "E:/GUVI VIDEOS/GUVI_PROJECT/PhonepePulseData8/data/aggregated/transaction/country/india/state/"

aggregated_transaction_list = os.listdir(path1)

column1 ={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[],"Transaction_amount":[] }

# access the state file
for state in aggregated_transaction_list:
    current_states = path1 + state + "/"
    aggregated_year_list = os.listdir(current_states)
    # access the year file
    for year in aggregated_year_list:
        current_years = current_states + year + "/"
        aggregated_file_list = os.listdir(current_years)
        # access the JSON file
        for file in aggregated_file_list:
            current_files = current_years+file
            data = open(current_files,"r")
            Value_data = json.load(data)
            # access the data
            for i in Value_data["data"]["transactionData"]:
                name = i["name"]
                count = i["paymentInstruments"][0]["count"]
                amount = i["paymentInstruments"][0]["amount"]
                # append the data to column1
                column1["Transaction_type"].append(name)
                column1["Transaction_count"].append(count)
                column1["Transaction_amount"].append(amount)
                column1["States"].append(state)
                column1["Years"].append(year)
                column1["Quarter"].append(int(file.strip(".json")))

aggregated_transaction = pd.DataFrame(column1)

aggregated_transaction["States"] = aggregated_transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggregated_transaction["States"] = aggregated_transaction["States"].str.replace("-"," ")
aggregated_transaction["States"] = aggregated_transaction["States"].str.title()
aggregated_transaction['States'] = aggregated_transaction['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

# print(aggregated_transaction)
```
